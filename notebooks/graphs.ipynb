```python
def plot_rewards(rewards, title='Rewards', xlabel='Timestep', ylabel='Reward'):
    mean_rewards = np.mean(rewards)
    timesteps = list(range(1, len(rewards) + 1))

    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    plt.plot(timesteps, rewards, label='Rewards')
    
    plt.axhline(y=mean_rewards, color='red', linestyle='--', linewidth=2, label='Mean Reward')

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Add legend
    plt.legend()
    plt.show()
```

```python
def plot_losses(losses, title='Losses', xlabel='Timestep', ylabel='Loss'):
    shifted_losses = losses[50:]
    timesteps = list(range(51, len(losses) + 1))

    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    plt.plot(timesteps, shifted_losses, label='Losses')

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Add legend
    plt.legend()
    plt.show()
```

```python
def plot_plastic_rewards(rewards, title='Rewards', xlabel='Timestep', ylabel='Reward'):
    fixed_len = int(0.7 * len(rewards))
    
    fixed_rewards = rewards[:fixed_len]
    plastic_rewards = rewards[fixed_len:]
    
    mean_fixed_reward = np.mean(fixed_rewards)
    mean_plastic_reward = np.mean(plastic_rewards)
    
    fixed_timesteps = list(range(1, fixed_len + 1))
    plastic_timesteps = list(range(fixed_len, len(rewards)))

    plt.figure(figsize=(10, 6))
    plt.plot(fixed_timesteps, fixed_rewards, label='Fixed Training Rewards', color="blue")
    plt.plot(plastic_timesteps, plastic_rewards, label="Plastic Training Rewards", color="orange")
    
    plt.plot([0, fixed_len], [mean_fixed_reward, mean_fixed_reward], linestyle='--', color='red', label="Mean Fixed Reward")
    plt.plot([fixed_len, len(rewards)], [mean_plastic_reward, mean_plastic_reward], linestyle='--', color='green', label="Mean Plastic Reward")
    
    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Add legend
    plt.legend()
    plt.show()
```

```python
def plot_multiple_rewards_normal(rewards, savepath, figsize=(10, 3), rows=1):
    fig, axs = plt.subplots(rows, int(math.ceil(len(rewards) / rows)), figsize=figsize)
    
    for i, reward in enumerate(rewards):
        current_rewards, env_name = reward
        x = range(1, len(current_rewards) + 1)
        
        fixed_len = int(0.7 * len(current_rewards))
    
        fixed_rewards = current_rewards[:fixed_len]
        plastic_rewards = current_rewards[fixed_len:]
    
        mean_fixed_reward = np.mean(fixed_rewards)
        mean_plastic_reward = np.mean(plastic_rewards)
    
        fixed_timesteps = list(range(1, fixed_len + 1))
        plastic_timesteps = list(range(fixed_len, len(current_rewards)))
        
        if rows == 1:
            axis = axs[i]
        else:
            row = int(i / rows)
            col = int(i % rows)
            axis = axs[row, col]
        
        current_mean_reward = np.mean(current_rewards)
        axis.plot(x, current_rewards, label='Rewards', color='blue')
        
        axis.plot([0, fixed_len], [mean_fixed_reward, mean_fixed_reward], linestyle='--', color='red', label="Mean Earlier Reward")
        axis.plot([fixed_len, len(current_rewards)], [mean_plastic_reward, mean_plastic_reward], linestyle='--', color='green', label="Mean Later Reward")
        
        axis.set_xlabel("Episode")
        axis.set_ylabel("Reward")
        axis.set_title(f"{env_name} Rewards")
        
    plt.tight_layout()
    plt.savefig(savepath)
    plt.show()
```

```python
plot_multiple_rewards_normal(
    [
        (breakout_dqn_rewards, "Breakout"),
        (seaquest_dqn_rewards, "SeqQuest"),
        (spaceinvaders_dqn_rewards, "SpaceInvaders")
    ],
    figsize=(10, 3),
    savepath="../data/outputs/dqn_rewards.png"
)
```

```python
def plot_items(items, savepath, figsize=(10, 3), rows=1):
    fig, axs = plt.subplots(rows, int(math.ceil(len(items) / rows)), figsize=figsize)
    
    for i, item in enumerate(items):
        current_item, start, offset, title, y_label = item
        x = range(start + offset + 1, start + len(current_item) + 1)
        shifted_items = current_item[offset:]
        
        if rows == 1:
            axis = axs[i]
        else:
            row = int(i / rows)
            col = int(i % rows)
            axis = axs[row, col]
        
        axis.plot(x, shifted_items, label=y_label, color='blue')
        
        axis.set_xlabel("Episode")
        axis.set_ylabel(y_label)
        axis.set_title(title)
        
    plt.tight_layout()
    plt.savefig(savepath)
    plt.show()
```
